```python
def str_concat(list):
    result = ''
    for element in list:
        result += ','+ str(element)
    return result
```

```python
def load_data_lst(my_url):
    data_lst =[]
    response = requests.get(my_url)
    jj = json.loads(response.text)
    results = jj['results']
    for result in results:
        name = result['name']
        types = str_concat(result['types']).strip(',')
        address = result['formatted_address']
        google_stars = result['rating']
        user_reviews = result['user_ratings_total']
        data = [name, types, address, google_stars, user_reviews]
        data_lst.append(data)
    return data_lst
```
