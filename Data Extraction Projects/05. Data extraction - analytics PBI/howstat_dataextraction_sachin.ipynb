# This notebook is created 
1. To extract Sachin Tendulkar's test match batting stats from "howstat.com"
2. Clean the extracted data and prepare a pandas dataframe
3. Export the pandas dataframe to csv for further visual analysis


```python
from urllib import request as req
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup as bs
import lxml
```

```python
import re
```

```python
url = Request('http://www.howstat.com/cricket/statistics/Players/PlayerBatGraph.asp?PlayerID=1735',headers={'User-Agent': 'Mozilla/5.0'})
url
```

```python
webpage = urlopen(url).read()
```

```python
page_soup = bs(webpage, 'lxml')
```

```python
page_soup
```

```python
all_data = page_soup.findAll("area")
```

```python
all_data[0]
```

```python
len(all_data)
```

```python
entries = []
```

```python
for i in range(len(all_data)):
    row_data = all_data[i]['title'].replace('\t','')#replacing \t with nothing
    row_data_1 = row_data.replace(' ','')#replacing space with nothing
    row_data_2 = row_data_1.replace('\r','')#replacing \r with nothing
    row_data_3 = row_data_2.replace('\n',' ') #replacing \n with space
    entries.append(row_data_3)
```

```python
entries
```

[re module](https://docs.python.org/3/library/re.html#making-a-phonebook)

```python
row_data_final = [re.split(":? ", entry, 6) for entry in entries]
row_data_final
```

```python
import pandas as pd
```

```python
headers = ['delete','Date','Runs Scored', 'Opponent','Ground','Aggregated Runs Scored','Aggregated Average']
```

```python
df = pd.DataFrame(data=row_data_final,columns=headers, index= list(range(1,330)))
```

```python
df
```

```python
df1 = df.drop(columns=['delete'])
df1
```

```python
df1['Date'] = df1['Date'].str.replace('Date:', '')
df1['Runs Scored'] = df1['Runs Scored'].str.replace('Runs:', '')
df1['Opponent'] = df1['Opponent'].str.replace('Versus:', '')
df1['Ground'] = df1['Ground'].str.replace('Ground:', '')
df1['Aggregated Runs Scored'] = df1['Aggregated Runs Scored'].str.replace('Aggregate:', '')
df1['Aggregated Average'] = df1['Aggregated Average'].str.replace('Average:', '')
df1
```

```python
df1['Year'] = [df1['Date'][i][-4:] for i in range(1,len(df1['Date'])+1) ]
```

```python
df1.to_csv('sachin_stat.csv')
```
